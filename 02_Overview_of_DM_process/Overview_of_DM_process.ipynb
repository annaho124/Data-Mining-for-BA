## DEAL WITH MISSING DATA

```python
#To illustrate missing data procedures, we first convert a few entries for bedroom to NA's.
#Then we impute these missing values using the median of the remaining values.
missingRows = wr.sample(10).index
wr.loc[missingRows, 'BEDROOMS'] = np.nan
print('Number of rows with valid BEDROOMS values after setting to NaN: ', wr['BEDROOMS'].count())
```

```python
#Remove rows with mising values
reduced_df = wr.dropna()
print('Number of rows after removing rows with missing values: ', len(reduced_df)) 
```

```python
#Replace the missing values using the median of the remaining values.
medianBedrooms = wr['BEDROOMS'].median()
wr.BEDROOMS = wr.BEDROOMS.fillna(value= medianBedrooms)
print('Number of rows with valid BEDROOMS values after filling NA values: ', wr['BEDROOMS'].count())
```

# 3. BUILDING A PREDICTIVE MODEL

## PREDICTIONS (FITTED VALUES) FOR A SAMPLE OF TRAING DATA

```python
from sklearn.linear_model import LinearRegression

#Create list of predictors and outcome
excludeColumns = ('TOTAL_VALUE', 'TAX')
predictors = [s for s in wr.columns if s not in excludeColumns]
outcome = 'TOTAL_VALUE'

#partition data
X = wr[predictors]
y = wr[outcome]
train_X, valid_X, train_y, valid_y = train_test_split(X, y, test_size = 0.4, random_state = 8)

model = LinearRegression()
model.fit(train_X, train_y)

train_pred = model.predict(train_X)
train_result = pd.DataFrame({
    'TOTAL_VALUE': train_y,
    'predicted'  : train_pred,
    'residual'   : train_y - train_pred
})
train_result.head()
```

##PREDICTIONS FOR A SAMPLE OF VALIDATION DATA

```python
valid_pred = model.predict(valid_X)
valid_result = pd.DataFrame({
    'TOTAL_VALUE'  : valid_y,
    'predicted'    : valid_pred,
    'residual'     : valid_y - valid_pred
})
valid_result.head()
```
